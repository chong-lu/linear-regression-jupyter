This notebook demonstrates a simple linear regression analysis using Python to model Salary based on Years of Experience.

```python
import pandas as pd
dataset = pd.read_csv("regression_data.csv")
```

```python
import matplotlib.pyplot as plt
plt.scatter(dataset[["YearsExperience"]], dataset[["Salary"]], color="red")
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(dataset[["YearsExperience"]], dataset[["Salary"]])
```

```python
plt.scatter(dataset[["YearsExperience"]], dataset[["Salary"]], color="red")
plt.plot(dataset["YearsExperience"], model.predict(dataset[["YearsExperience"]]), color="blue")
plt.title("Salary vs Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
model.score(dataset[["YearsExperience"]], dataset[["Salary"]])  # R-squared
```

```python
from scipy.stats import linregress
slope, intercept, r_value, p_value, std_err = linregress(dataset["YearsExperience"], dataset["Salary"])
```

```python
linregress(dataset["YearsExperience"], dataset["Salary"])
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(dataset["Salary"], 
                         model.predict(dataset[["YearsExperience"]]))
mse
```

```python
plt.scatter(dataset[["YearsExperience"]], dataset[["Salary"]], color="red")
plt.plot(dataset["YearsExperience"], model.predict(dataset[["YearsExperience"]]), label='Fitted line', color="blue")
plt.title("Salary vs Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.text(1.25, 60000,f"y = {slope:.2f}x + {intercept:.2f}\nr = {r_value:.2f}")
plt.show()
```
